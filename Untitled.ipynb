```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv('ma_lga_12345.csv', parse_dates=["saledate"], dayfirst=True)
dataset = dataset.set_index("saledate").to_period("M").reset_index()
dataset.head()
```

```python
new_dataset = dataset.set_index("saledate")
type_bed_grp = new_dataset.groupby(["type", "bedrooms"])
house_2_data = type_bed_grp.get_group(("house", 2))
house_2_data["Time"] = np.arange(len(house_2_data))
house_2_data = house_2_data.reset_index()
```

```python
house_2_data.head()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(7, 1, figsize=(6,20))

counter = 0
for i in type_bed_grp.groups.keys():
    data = type_bed_grp.get_group(i)
    data["Time"] = np.arange(len(data))
    data = data.reset_index()
    ax[counter].scatter('Time', 'MA', data=data)
    sns.regplot(x="Time", y="MA", data=data, scatter_kws={"color": "blue"}, color="red", ax=ax[counter])
    counter = counter + 1
```
